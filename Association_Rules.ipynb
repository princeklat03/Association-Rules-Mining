## Loading the data and Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df_assoc = pd.read_excel("/Users/PKLAT/Desktop/Edureka/week 5/Online_Retail.xlsx")
```

```python
df_assoc.head()
```

### Cleaning the data

```python
df_assoc['Description'] = df_assoc['Description'].str.strip(" ")
```

```python
df_assoc.drop('CustomerID', axis=1, inplace=True)
```

```python
df_assoc2 = df_assoc[~df_assoc['InvoiceNo'].str.contains('C', na=False)] #Taking the invoices that have been bought by customer; identified by invoices that do not contain 'C' in it
```

```python
df_assoc2.head()
```

### Subsampling the data to do analysis only for France

```python
df_assoc2_France = df_assoc2[df_assoc2.Country == 'France']
```

```python
df_assoc2_France.head()
```

```python
df_assoc2_France.info()
```

```python
basket = df_assoc2_France.groupby(['InvoiceNo', 'Description'])
```

```python
basket.get_group((536370, "ALARM CLOCK BAKELIKE RED"))
```

```python
basket_real = basket['Quantity'].sum()
```

```python
basket_real
```

```python
basket_df = basket_real.unstack()
```

```python
basket_df
```

```python
basket_df.fillna(0, inplace=True)
```

```python
basket_df.head(50)
```

```python
def encode(x):
    if(x>0):
        return 1
    else:
        return 0
```

```python
basket_df2 = basket_df.applymap(encode)
```

```python
basket_df2.head(50)
```

```python
basket_df2.drop('POSTAGE', axis=1, inplace=True)
```

#### Applying apriori algo

```python
basket_apriori = apriori(basket_df2, min_support=0.07, use_colnames=True)
```

```python
basket_apriori
```

```python
association_rul = association_rules(basket_apriori, metric='lift',min_threshold=1)
```

```python
association_rul
```
